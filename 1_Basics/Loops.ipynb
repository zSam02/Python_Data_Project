```python
#While Loops - Checked condition of TRUE and FALSE continues only if TRUE
```

```python
count = 1
while count < 5:
  print(count)
  count += 1
```

```python
years = 0
index = 0


years_list = list(years_experience.items())
while years < 5:
  key, years = years_list[index]
  if years >=5:
    print(key, 'has', years, 'years of experience.')
  index += 1
```
