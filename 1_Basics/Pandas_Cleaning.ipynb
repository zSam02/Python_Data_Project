```python
# Install datasets Library (if not already installed)
!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
# Convert 'job_posted_date' to datetime without specifying the exact format
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')

df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
df.head()
```

```python
df.sort_values(by='job_posted_date', ascending=False, inplace=True) # Sorts value in order
df
```
