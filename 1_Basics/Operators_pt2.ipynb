```python
def solution(N):
    # Convert the number to its binary representation (excluding the '0b' prefix)
    binary_rep = bin(N)[2:]

    # Split the binary string at '1's and filter out empty gaps (trailing zeros aren't valid gaps)
    gaps = binary_rep.strip('0').split('1')

    # Return the length of the longest binary gap or 0 if there are no gaps
    return max(map(len, gaps), default=0)

# Test the function with the example cases
test_cases = [9, 529, 20, 15, 32, 1041]
results = {n: solution(n) for n in test_cases}
results

```

```python
temperatures = [0, 1, 6, -2, -2, 1, 2, 2]

def negative(temperatures):
  days = 0

  for x in temperatures:
    if x < 0:
      days += 1
  return days

days = negative(temperatures) # Assign the returned value from the function negative(temperatures) to the variable days
print(f"The number of days in which temperature is negative is: {days}") # Print the output with the requested text.
```

```python
def solution(A):
    unpaired = 0

    for number in A:
        unpaired ^= number

    return unpaired
```
