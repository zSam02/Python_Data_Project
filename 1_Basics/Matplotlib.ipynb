```python
# Install datasets Library (if not already installed)
#!pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
date_counts = df.job_posted_date.value_counts()

date_counts = date_counts.sort_index()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month #we aggregate data by months
```

```python
df
```

```python
# Group by 'month' and count the job postings
monthly_counts = df['job_posted_month'].value_counts()

# Sort the months
monthly_counts = monthly_counts.sort_index()

monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts.values)
plt.show()
```

```python
#January has a surge in job postings
```
