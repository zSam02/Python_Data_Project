# Pandas Index Managment 

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(3)
```

```python
df.index
```

```python
df.index.name = 'job_index' #names the first column(0)
```

```python
df.head()
```

```python
df.index.dtype
```

# reset_index()

```python
df_uk = df[df['job_country']== 'United Kingdom']
```

```python
df_uk.head() # job index has missing values, so we reset the index 
```

```python
df_uk.reset_index(inplace=True)
df_uk
```

# set_index()

```python
df_uk.set_index('job_index', inplace=True) # loses data 
```

```python
df_uk
```

# sort_index

```python
median_pivot = df_uk.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True) # values are sorted by the index and in alphabetical order
```

```python
median_pivot # sort index methods workds well when there is no name for the index column 
```
