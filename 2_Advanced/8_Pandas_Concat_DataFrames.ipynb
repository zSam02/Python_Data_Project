```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b') # Lower case give first three letters 

df
```

```python
# We want a dictionary of all the months 

months = df['job_posted_month'].unique() # gives unique values


#Write this with dict comprehension
dict_months = {month: df[df['job_posted_month'] == month] for month in months}

type(dict_months)
```
