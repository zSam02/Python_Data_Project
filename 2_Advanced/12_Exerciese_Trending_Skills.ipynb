```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
 # create a new dataframe setting it to our original filtering for job_title_short ensuring it equals Data Analyst
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month # aggregating skills on a monthly basis 
```

```python
df_DA_explode= df_DA.explode('job_skills') #Explode the job_skills column. with new data frame jobs_skills seperated out 
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills',  aggfunc='size', fill_value=0) #size counts the job posts but has NaN values so se use fill values 

df_DA_pivot
```

```python
# we need to create a row that sum all values as there are to many skills 
# Create a total row
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot
```

```python
# Order columns by total count
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
# Drop total row and sort by job skills 
df_DA_pivot = df_DA_pivot.drop('Total') # drops the total row 

df_DA_pivot
```

```python
# code snippet from `6_Exercise_Job_Demand.ipynb` I like to have my months as names, not numbers, so let's change this.
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
